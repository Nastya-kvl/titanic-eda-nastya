```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загружаем датасет
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Смотрим первые строки
df.head()
```

```python
# Размер таблицы
print("Форма таблицы:", df.shape)

# Типы данных + пропуски
df.info()
```

```python
# Статистика по числовым колонкам
df.describe()
```

```python
# Сумма пропущенных значений по каждой колонке
df.isnull().sum()
```

```python
sns.countplot(x='Sex', data=df)
plt.title('Пол пассажиров на борту')
plt.show()
```

```python
sns.histplot(data=df, x='Age', kde=True)
plt.title('Распределение по возрасту')
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Кто выжил')
plt.xticks([0, 1], ['Нет', 'Да'])  # Подписи на осях
plt.show()
```

## Выводы по анализу данных Titanic

- В датасете **891 пассажир** и **12 признаков**
- Есть пропущенные значения в колонках `Age`, `Cabin`, `Embarked`
- На борту было **больше мужчин**, чем женщин
